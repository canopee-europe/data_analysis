```python
import requests
import pandas as pd
from io import StringIO
```

```python
url = ("https://static.data.gouv.fr/resources/communes-de-france-base-des-codes-postaux/20241113-073516/20230823-communes-departement-region.csv")
response = requests.get(url)
```

```python
# Put data into df
if response.status_code == 200:
    data = response.text
    df_communes_zipcode = pd.read_csv(StringIO(data))
else:
    print("Erreur")

df_communes_zipcode
```

```python
# Keep usefull columns
df_communes_zipcode_final = df_communes_zipcode[['code_postal', 'nom_commune_complet', 'code_departement', 'nom_departement', 'nom_region']]
df_communes_zipcode_final
```

```python
# Drop rows
index_to_drop = df_communes_zipcode_final[df_communes_zipcode_final['nom_departement'].isnull()].index
df_communes_zipcode_final.drop(index_to_drop, inplace=True)
df_communes_zipcode_final
```

```python
df_communes_zipcode_final.drop_duplicates(subset=['nom_commune_complet'], inplace=True)
df_communes_zipcode_final
```

```python
df_communes_zipcode_final['code_postal'] = df_communes_zipcode_final['code_postal'].astype(str).str.zfill(5)
df_communes_zipcode_final
```

```python
df_communes_zipcode_final['code_departement'] = df_communes_zipcode_final['code_departement'].astype(str).str.zfill(2)
df_communes_zipcode_final
```

```python
df_communes_zipcode_final.to_csv('correspondance_zipcode_commune.csv', index=False)
```
