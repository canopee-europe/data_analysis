```python
import pandas as pd
import requests
import config
```

## Extract

```python
membership_url ="https://dashboard.activcreew.app/api/memberships?fields=mb_status, role&populate[association][fields]=id&populate[user][fields]=id"
response_membership = requests.get(membership_url, headers=config.my_headers)
membership_data = response_membership.json()["data"]
df_membership = pd.DataFrame(membership_data)
df_membership
```

## Transform

```python
# Get fields from columns dicts
df_membership["association_id"] = df_membership["association"].apply(lambda x: x.get("id"))
df_membership["user_id"] = df_membership["user"].apply(lambda x: x.get("id"))
df_membership
```

```python
# Drop useless columns
df_membership.drop(columns=["documentId", 'association', 'user'], inplace=True)
df_membership
```

## Load

```python
df_membership.to_csv("csv_files/membership_data.csv", index=False)
```
