```python
import pandas as pd
import requests
import config
```

## Extract

```python
membership_url ="https://dashboard.activcreew.app/api/memberships?fields=mb_status, role&populate[association][fields]=id&populate[user][fields]=id&pagination[pageSize]=500"
page = 1
list_membership_data = []

while True:
    params = {
        "pagination[page]": page,
    }
    response_membership = requests.get(membership_url, headers=config.my_headers, params=params)
    membership_data = response_membership.json()
    list_membership_data.extend(membership_data["data"])

    page_count = membership_data.get("meta", {}).get("pagination", {}).get("pageCount", 1)

    if page >= page_count:
        break
    page += 1

print(len(list_membership_data))
```

## Transform

```python
df_membership = pd.DataFrame(list_membership_data)
df_membership
```

```python
# Get fields from columns dicts
df_membership["association_id"] = df_membership["association"].apply(lambda x: x.get("id") if isinstance(x, dict) else None)
df_membership["user_id"] = df_membership["user"].apply(lambda x: x.get("id"))
df_membership
```

```python
# Drop useless columns
df_membership.drop(columns=["documentId", 'association', 'user'], inplace=True)
df_membership
```

## Load

```python
df_membership.to_csv("csv_files/membership_data.csv", index=False)
```
