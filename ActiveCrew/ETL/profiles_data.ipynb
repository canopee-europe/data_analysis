```python
import pandas as pd
import requests
from datetime import datetime

import config
```

## Extract

```python
profiles_url = "https://dashboard.activcreew.app/api/profiles?fields=gender, birthday, birthplace"
response_profiles = requests.get(profiles_url, headers=config.my_headers)
profiles_data = response_profiles.json()["data"]
df_profiles = pd.DataFrame(profiles_data)
df_profiles
```

```python
df_profiles.dtypes
```

## Transform

```python
df_profiles.drop(columns='documentId', inplace=True)
df_profiles
```

```python
df_profiles['birthday'] = pd.to_datetime(df_profiles['birthday'])
df_profiles
```

```python
df_profiles['age'] = df_profiles['birthday'].apply(lambda x: (datetime.now() - x).days // 365)
df_profiles
```
