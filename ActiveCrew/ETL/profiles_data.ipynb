```python
import pandas as pd
import requests
from datetime import datetime

import config
```

## Extract

```python
profiles_url = "https://dashboard.activcreew.app/api/profiles?fields=gender, birthday, birthplace&populate[addresses][fields]=city, country&pagination[pageSize]=500"

page = 1
list_profiles_data = []

while True:
    params = {
        "pagination[page]": page,
    }
    response_profiles = requests.get(profiles_url, headers=config.my_headers, params=params)
    profiles_data = response_profiles.json()
    list_profiles_data.extend(profiles_data["data"])

    page_count = profiles_data.get("meta", {}).get("pagination", {}).get("pageCount", 1)

    if page >= page_count:
        break
    page += 1
```

## Transform

```python
df_profiles = pd.DataFrame(list_profiles_data)
df_profiles
```

```python
df_profiles["city_of_residence"] = df_profiles["addresses"].apply(lambda x: x[0]['city'])
df_profiles["country_of_residence"] = df_profiles["addresses"].apply(lambda x: x[0]['country'])
df_profiles
```

```python
df_profiles.drop(columns=['documentId', 'addresses'], inplace=True)
df_profiles
```

```python
df_profiles['birthday'] = pd.to_datetime(df_profiles['birthday'])
df_profiles
```

```python
df_profiles['age'] = df_profiles['birthday'].apply(lambda x: (datetime.now() - x).days // 365)
df_profiles
```

## Load

```python
df_profiles.to_csv('csv_files/profiles_data.csv', index=False)
```
