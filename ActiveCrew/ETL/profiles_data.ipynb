```python
import pandas as pd
import requests
from datetime import datetime

import config
```

## Extract

```python
profiles_url = "https://dashboard.activcreew.app/api/profiles?fields=gender, birthday&populate[addresses][fields]=city, country, zipcode&pagination[pageSize]=500"

page = 1
list_profiles_data = []

while True:
    params = {
        "pagination[page]": page,
    }
    response_profiles = requests.get(profiles_url, headers=config.my_headers, params=params)
    profiles_data = response_profiles.json()
    list_profiles_data.extend(profiles_data["data"])

    page_count = profiles_data.get("meta", {}).get("pagination", {}).get("pageCount", 1)

    if page >= page_count:
        break
    page += 1
```

## Transform

```python
list_profiles_data_copy = list_profiles_data.copy()
list_profiles_data_copy
```

```python
df_profiles = pd.DataFrame(list_profiles_data_copy)
df_profiles
```

```python
# Get fields from columns
df_profiles["city_of_residence"] = df_profiles["addresses"].apply(lambda x: x[0]['city'])
df_profiles["country_of_residence"] = df_profiles["addresses"].apply(lambda x: x[0]['country'])
df_profiles["zipcode"] = df_profiles["addresses"].apply(lambda x: x[0]['zipcode'])
df_profiles
```

```python
# Drop useless columns
df_profiles.drop(columns=['documentId', 'addresses'], inplace=True)
df_profiles
```

```python
# Convert birthday to datetime
def fix_date_string(date_str):
    try:
        parts = date_str.split('-')
        year = int(parts[0])
        # Correction des ann√©es sur 3 chiffres
        if year < 1000:
            year += 1900
        return f"{year:04d}-{parts[1]}-{parts[2]}"
    except Exception as e:
        return None  # ou date_str si tu veux conserver la valeur d'origine


df_profiles['birthday'] = df_profiles['birthday'].apply(fix_date_string)


try :
    df_profiles['birthday'] = pd.to_datetime(df_profiles['birthday'])
except Exception as e:
    invalid_dates = []
    for val in df_profiles['birthday']:
        try:
            pd.to_datetime(val)
        except Exception:
            invalid_dates.append(val)
    print("Les valeurs invalides sont :", invalid_dates)
```

```python
df_profiles
```

```python
# Calculate age profile
df_profiles['age'] = df_profiles['birthday'].apply(lambda x: (datetime.now() - x).days // 365)
df_profiles
```

```python
df_profiles_copy = df_profiles.copy()
```

```python
df_profiles_copy
```

```python
df_profiles_copy = df_profiles_copy.drop(df_profiles_copy[df_profiles_copy['zipcode'] == ''].index)
df_profiles_copy
```

```python
df_profiles_copy['zipcode'] = df_profiles_copy['zipcode'].str.extract(r'(\d+)', expand=False)
df_profiles_copy
```

```python
has_digit = df_profiles_copy['zipcode'].str.contains(r'\d', na=False)
has_alpha = df_profiles_copy['zipcode'].str.contains(r'[a-zA-Z]', na=False)
```

```python
df_profiles_copy.loc[has_alpha, 'zipcode']
```

```python
# Fonction pour remplir les NaN depuis 'city'
def fill_zipcode_from_city(row):
    if pd.isna(row['zipcode']):
        if isinstance(row['city_of_residence'], str):
            match = ''.join(filter(str.isdigit, row['city_of_residence']))
            if match:
                return match
    return row['zipcode']

# Application ligne par ligne
df_profiles_copy['zipcode'] = df_profiles_copy.apply(fill_zipcode_from_city, axis=1)

# Conversion finale en entier nullable
# df_profiles_copy['zipcode_clean'] = pd.to_numeric(df_profiles_copy['zipcode_clean'], errors='coerce').astype('Int64')

df_profiles_copy
```

```python
df_ready = df_profiles_copy.copy()
```

```python
# Chargement du r√©f√©rentiel CSV
referentiel = pd.read_csv("communes/correspondance_zipcode_commune.csv", dtype={"code_postal": str})

# Extraction des 2 premiers chiffres du code postal dans le r√©f√©rentiel
referentiel["prefix_cp"] = referentiel["code_postal"].str[:2]

# On garde uniquement la premi√®re occurrence par code d√©partement (pr√©fixe CP)
departement_info = referentiel.drop_duplicates(subset=["prefix_cp"])[["prefix_cp", "nom_departement", "nom_region"]]

departement_info
```

```python
# Chargement de ton DataFrame principal (ex: avec la colonne 'zipcode')
# Assure-toi que 'zipcode' est bien une cha√Æne de caract√®res
df_ready["zipcode"] = df_ready["zipcode"].astype(str)

# Extraction des deux premiers chiffres
df_ready["dept_candidate"] = df_ready["zipcode"].str[:2]
df_ready
```

```python
# Jointure entre df et le r√©f√©rentiel bas√© sur les 2 premiers chiffres
df_ready = df_ready.merge(departement_info, how="left", left_on="dept_candidate", right_on="prefix_cp")

df_ready
```

```python
# Nettoyage : suppression de la colonne interm√©diaire si souhait√©
df_ready = df_ready.drop(columns=["prefix_cp"])
# df_ready = df_ready.drop(columns=["dept_candidate"])

df_ready
```
