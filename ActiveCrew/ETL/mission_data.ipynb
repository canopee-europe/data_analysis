```python
import pandas as pd
import requests
import config
```

## Extract

```python
mission_url = "https://dashboard.activcreew.app/api/missions?fields=id, title&populate[event][fields]=id&pagination[pageSize]=500"

page = 1
list_mission_data = []

while True:
    params = {
        "pagination[page]": page,
    }
    response_mission = requests.get(mission_url, headers=config.my_headers, params=params)
    missions_data = response_mission.json()
    list_mission_data.extend(missions_data["data"])

    page_count = missions_data.get("meta", {}).get("pagination", {}).get("pageCount", 1)

    if page >= page_count:
        break
    page += 1
```

## Transform

```python
df_missions = pd.DataFrame(list_mission_data)
df_missions
```

```python
# Get field from columns dicts
df_missions["event_id"] = df_missions["event"].apply(lambda x: x.get("id") if isinstance(x, dict) else None)
df_missions
```

```python
index_null_event = df_missions[df_missions["event_id"].isnull()].index
df_missions.drop(index_null_event, inplace=True)
df_missions
```

```python
# Type conversions
convert_dict = {"event_id": int}
df_missions = df_missions.astype(convert_dict)

df_missions
```

```python
# Drop useless columns
df_missions.drop(columns=["documentId", 'event'], inplace=True)
df_missions
```

## Load

```python
df_missions.to_csv("csv_files/mission_data.csv", index=False)
```
