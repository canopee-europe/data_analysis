```python
import pandas as pd
import requests
import config
```

## Extract

```python
mission_url = "https://dashboard.activcreew.app/api/missions?fields=id, title&populate[event][fields]=id"
response_mission = requests.get(mission_url, headers=config.my_headers)
missions_data = response_mission.json()["data"]
df_missions = pd.DataFrame(missions_data)
df_missions
```

## Transform

```python
# Get field from columns dicts
df_missions["event_id"] = df_missions["event"].apply(lambda x: x.get("id") if isinstance(x, dict) else None)
df_missions
```

```python
# Fill na
df_missions["event_id"].fillna(0, inplace=True)

# Type conversions
convert_dict = {"event_id": int}
df_missions = df_missions.astype(convert_dict)

df_missions
```

```python
# Drop useless columns
df_missions.drop(columns=["documentId", 'event'], inplace=True)
df_missions
```
