```python
import pandas as pd
import requests
import config
```

## Extract

```python
mission_url = "https://dashboard.activcreew.app/api/missions?fields=id, title&populate[prefix_mission][fields]=name"

page = 1
list_mission_data = []

while True:
    params = {
        "pagination[page]": page,
    }
    response_mission = requests.get(mission_url, headers=config.my_headers, params=params)
    missions_data = response_mission.json()
    list_mission_data.extend(missions_data["data"])

    page_count = missions_data.get("meta", {}).get("pagination", {}).get("pageCount", 1)

    if page >= page_count:
        break
    page += 1
```

## Transform

```python
df_missions = pd.DataFrame(list_mission_data)
df_missions
```

```python
# Get field from columns dicts
df_missions["prefix_mission_name"] = df_missions["prefix_mission"].apply(lambda x: x.get("name") if isinstance(x, dict) else None)
df_missions
```

```python
# Drop useless columns
df_missions.drop(columns=["documentId", 'prefix_mission'], inplace=True)
df_missions
```

## Load

```python
df_missions.to_csv("csv_files/mission_data.csv", index=False)
```

## Extract

```python
event_edition_mission_url ="https://dashboard.activcreew.app/api/event-editions?fields=name&populate[missions][fields]=members&pagination[pageSize]=500"
response = requests.get(event_edition_mission_url, headers=config.my_headers)
event_edition_data = response.json()["data"]
df_event_edition_mission = pd.DataFrame(event_edition_data)
df_event_edition_mission
```

## Transform

```python
df_event_edition_mission = df_event_edition_mission.explode('missions')
```

```python
df_event_edition_mission['mission_id'] = df_event_edition_mission['missions'].apply(lambda x: x.get("id") if isinstance(x, dict) else None)
df_event_edition_mission['members'] = df_event_edition_mission['missions'].apply(lambda x: x.get("members")if isinstance(x, dict) else None)
df_event_edition_mission
```

```python
df_event_edition_mission.drop(columns=["documentId", "missions"], inplace=True)
df_event_edition_mission
```

```python
df_merged = df_missions.merge(df_event_edition_mission, left_on="id", right_on="mission_id", how="left")
df_merged
```

```python
df_merged = df_merged.rename(columns={"id_x": "id_mission", "id_y": "event_edition_id"})
df_merged
```

```python
df_merged.drop(columns="mission_id", inplace=True)
df_merged
```

```python
df_merged.to_csv("csv_files/event_edition_mission_data.csv", index=False)
```
