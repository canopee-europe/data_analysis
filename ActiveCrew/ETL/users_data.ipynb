```python
import pandas as pd
import requests
import config
```

## Extract

```python
user_url = "https://dashboard.activcreew.app/api/users?fields=id&populate[profile][fields]=id&pagination[pageSize]=500"
response_user = requests.get(user_url, headers=config.my_headers)
user_data = response_user.json()
df_user = pd.DataFrame(user_data)
df_user
```

## Transform

```python
# Get fields from columns dicts
df_user["profile_id"] = df_user["profile"].apply(lambda x: x.get("id") if isinstance(x, dict) else None)
df_user
```

```python
# Fill na
df_user["profile_id"].fillna(0, inplace=True)
```

```python
# Type conversions
convert_dict = {"profile_id": int}
df_user = df_user.astype(convert_dict)
df_user
```

```python
# Drop useless columns
df_user.drop(columns=["profile", "documentId"], inplace=True)
df_user
```

```python
index_to_remove = df_user[df_user["profile_id"] == 0].index
df_user.drop(index_to_remove, inplace=True)
df_user
```

## Load

```python
df_user.to_csv("csv_files/users_data.csv", index=False)
```
