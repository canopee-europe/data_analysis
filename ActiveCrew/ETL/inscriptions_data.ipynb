```python
import pandas as pd
import requests
import datetime
import config
```

## Extract

```python
inscription_url = "https://dashboard.activcreew.app/api/inscriptions?fields=id, createdAt&populate[event_edition][fields]=id&populate[user][fields]=id&pagination[pageSize]=500"
page = 1
list_inscription_data = []

while True:
    params = {
        "pagination[page]": page,
    }
    response_inscription = requests.get(inscription_url, headers=config.my_headers, params=params)
    inscriptions_data = response_inscription.json()
    list_inscription_data.extend(inscriptions_data["data"])

    page_count = inscriptions_data.get("meta", {}).get("pagination", {}).get("pageCount", 1)

    if page >= page_count:
        break
    page += 1
```

## Transform

```python
df_inscription = pd.DataFrame(list_inscription_data)
df_inscription
```

```python
# Get fields from columns dicts
df_inscription["event_edition_id"] = df_inscription["event_edition"].apply(lambda x: x.get("id"))
df_inscription["user_id"] = df_inscription["user"].apply(lambda x: x.get("id"))
df_inscription
```

```python
# Drop useless columns
df_inscription.drop(columns=["event_edition", "documentId", "user"], inplace=True)
df_inscription
```

```python
# Convert createdAt to datetime and format output
df_inscription["createdAt"] = pd.to_datetime(df_inscription["createdAt"])
df_inscription["createdAt"] = df_inscription["createdAt"].dt.strftime("%Y-%m-%d")
df_inscription
```

## Load

```python
df_inscription.to_csv('csv_files/inscriptions_data.csv', index=False)
```
