```python
import pandas as pd
import requests
import datetime
import config
```

## Extract

```python
inscription_url = "https://dashboard.activcreew.app/api/inscriptions?fields=id, createdAt&populate[event_edition][fields]=id"
response_inscription = requests.get(inscription_url, headers=config.my_headers)
inscriptions_data = response_inscription.json()["data"]
df_inscription = pd.DataFrame(inscriptions_data)
df_inscription
```

## Transform

```python
# Get fields from columns dicts
df_inscription["event_edition_id"] = df_inscription["event_edition"].apply(lambda x: x.get("id"))
df_inscription
```

```python
# Drop useless columns
df_inscription.drop(columns=["event_edition", "documentId"], inplace=True)
df_inscription
```

```python
# Convert createdAt to datetime and format output
df_inscription["createdAt"] = pd.to_datetime(df_inscription["createdAt"])
df_inscription["createdAt"] = df_inscription["createdAt"].dt.strftime("%Y-%m-%d")
df_inscription
```

## Load

```python
df_inscription.to_csv('csv_files/inscriptions_data.csv', index=False)
```
