```python
import pandas as pd
import requests
import config
```

## Extract

```python
associations_url = "https://dashboard.activcreew.app/api/associations?fields=id, name&populate[event][fields]=id, name&populate[address][fields]=id, country, city&pagination[pageSize]=500"
response_assos = requests.get(associations_url, headers=config.my_headers)
data_assos = response_assos.json()["data"]
df_assos = pd.DataFrame(data_assos)
df_assos
```

## Transform

```python
# Get event_id
df_assos["event_id"] = df_assos["event"].apply(lambda x: x.get("id") if isinstance(x, dict) else None)
df_assos["event_name"] = df_assos["event"].apply(lambda x: x.get("name") if isinstance(x, dict) else None)

# Get country
df_assos["country"] = df_assos["address"].apply(lambda x: x.get("country") if isinstance(x, dict) else None)
df_assos["city"] = df_assos["address"].apply(lambda x: x.get("city") if isinstance(x, dict) else None)
df_assos
```

```python
# Drop columns
df_assos.drop(columns=["event", "address", "documentId"], inplace=True)
df_assos
```

```python
# Fill na
df_assos["event_id"].fillna(0, inplace=True)

# Type conversions
convert_dict = {"event_id": int}
df_assos = df_assos.astype(convert_dict)


df_assos
```

```python
# Drop rows where event_id == 0
index_null = df_assos[df_assos['event_id'] == 0].index
df_assos.drop(index_null, inplace=True)
df_assos
```

```python
# Normalize city field
df_assos['city'] = df_assos['city'].str.lower()
df_assos['city'] = df_assos['city'].str.capitalize()
df_assos
```

```python
# Fill Na country
df_assos['country'] = df_assos['country'].fillna('FR')
df_assos
```

## Load

```python
df_assos.to_csv('csv_files/associations_data.csv', index=False)
```
