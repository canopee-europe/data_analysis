```python
import pandas as pd
import requests
import config
```

## Extract

```python
asso_inscription_url = "https://dashboard.activcreew.app/api/asso-inscriptions?fields=event_accepted, asso_accepted&populate[event_edition][fields]=id, name&populate[users][fields]=id&&pagination[pageSize]=500"

response_ai = requests.get(asso_inscription_url, headers=config.my_headers)
ai_data = response_ai.json()["data"]
df_asso_inscriptions = pd.DataFrame(ai_data)
df_asso_inscriptions
```

## Transform

```python
df_asso_inscriptions["event_edition_id"] = df_asso_inscriptions["event_edition"].apply(lambda x: x.get('id'))
df_asso_inscriptions["event_edition_name"] = df_asso_inscriptions["event_edition"].apply(lambda x: x.get('name'))
df_asso_inscriptions
```

```python
df_asso_inscriptions.drop(columns='event_edition', inplace=True)
df_asso_inscriptions
```

```python
df_asso_inscriptions_copy = df_asso_inscriptions.copy()
```

```python
df_asso_inscriptions_copy['user_ids'] = df_asso_inscriptions_copy['users'].apply(lambda user_list: [d['id'] for d in user_list])
df_asso_inscriptions_copy
```

```python
df_asso_inscriptions_copy.drop(columns=['users', 'documentId'], inplace=True)
df_asso_inscriptions_copy
```

```python
df_asso_inscriptions_copy['nb_users'] = df_asso_inscriptions_copy['user_ids'].apply(len)
df_asso_inscriptions_copy
```

```python
df_asso_inscriptions_copy.drop(columns='user_ids', inplace=True)
df_asso_inscriptions_copy
```

## Load

```python
df_asso_inscriptions_copy.to_csv('csv_files/asso_inscriptions_data.csv', index=False)
```
