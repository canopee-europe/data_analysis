```python
import pandas as pd
import requests
import config
```

## Extract

```python
asso_inscription_url = "https://dashboard.activcreew.app/api/asso-inscriptions?fields=event_accepted, asso_accepted&populate[event_edition][fields]=id&populate[users][fields]=id&&pagination[pageSize]=500"

response_ai = requests.get(asso_inscription_url, headers=config.my_headers)
ai_data = response_ai.json()["data"]
df_asso_inscriptions = pd.DataFrame(ai_data)
df_asso_inscriptions
```

## Transform

```python
df_asso_inscriptions["event_edition_id"] = df_asso_inscriptions["event_edition"].apply(lambda x: x.get('id'))
df_asso_inscriptions
```
