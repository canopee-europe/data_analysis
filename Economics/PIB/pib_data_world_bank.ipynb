```python
import pandas as pd
import requests
```

### Import DATA

```python
df = pd.read_csv('data/WB_WDI_NY_GDP_MKTP_CD_WIDEF.csv')
df
```

### Prepare dataframe

```python
df = df.rename(columns={'REF_AREA_LABEL': 'Pays'})
df
```

```python
df_clean = df_long.dropna(subset=['Value'])
df_clean
```

```python
df_clean['Value'] = df_clean['Value'] / 1e9
df_clean
```

```python
df_clean = df_clean.rename(columns={'Value': 'Value (en Mds)'})
df_clean
```

```python
df_clean.to_csv('Historic_GDP_per_country.csv', index=False)
```

## Calculation

### Calculate annual variation by country

```python
df_annual_variation = df_filtered.sort_values(["Pays", "Year"])
df_annual_variation
```

```python
df_annual_variation["Variation_%"] = (df_annual_variation.groupby("Pays")["Value (en Mds)"].pct_change() * 100)
df_annual_variation
```

#### Cumsum value by country

```python
df_annual_variation["Cumsum value"] = df_annual_variation.groupby("Pays")["Value (en Mds)"].cumsum()
df_annual_variation["Year"] = df_annual_variation["Year"].astype(int)
df_annual_variation
```

```python
df_annual_variation.to_csv('historic_annual_variation_g20.csv', index=False)
```

#### Calculate time to double value

```python
def doubling_times(df_country):
    """
    Calcule les temps de doublement pour un seul pays.
    Retourne une liste de tuples (année_départ, année_doublage, durée).
    """
    results = []
    values = df_country.sort_values("Year")[["Year", "Value (en Mds)"]].to_numpy()

    start_year, start_value = values[0]
    target = 2 * start_value

    for year, value in values[1:]:
        if value >= target:
            duration = int(year) - int(start_year)
            results.append((int(start_year), int(year), duration, start_value, value))
            # Nouveau point de départ
            start_year, start_value = year, value
            target = 2 * start_value

    return results
```

```python
doubling_all = []

for country, group in df_filtered.groupby("Pays"):
    records = doubling_times(group)
    for start_year, end_year, duration, start_value, value in records:
        doubling_all.append({
            "Pays": country,
            "Année_départ": start_year,
            "Value_start": start_value,
            "Année_doublage": end_year,
            "Value_end": value,
            "Dates": (start_year, end_year),
            "Durée": duration,
            "Values": (start_value, value)
        })
doubling_all
```

```python
df_doubled_gdp = pd.DataFrame(doubling_all)
df_doubled_gdp
```

```python
df_doubled_gdp.to_csv('doubled_gdp_per_country.csv', index=False)
```
