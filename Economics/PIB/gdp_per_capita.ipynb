```python
import pandas as pd
```

```python
df = pd.read_csv('data/historic_gdp_per_country.csv')
df
```

```python
df.drop(columns=["Country Code", 'Indicator Code', 'Indicator Name'], inplace=True)
df
```

```python
df_long = df.melt(id_vars=["Country Name"], var_name="Year", value_name="Value")
df_long
```

```python
df_long = df_long.dropna(subset=['Value'])
df_long
```

```python
# Set list of countries in G20
list_countries_g20 = ["Germany", "France", "Italy", "Argentina", "Australia", "Brazil", "Canada", "China",
"India", "Indonesia", "Japan", "Mexico", "Russian Federation", "Saudi Arabia", "South Africa", "South Korea", "Turkiye",
"United Kingdom", "United States"]
```

```python
df_filtered = df_long[df_long["Country Name"].isin(list_countries_g20)]
df_filtered
```

```python
df_filtered.to_csv('output/gdp_per_capita.csv', index=False)
```
