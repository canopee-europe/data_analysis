```python
import pandas as pd
```

```python
df = pd.read_csv('historic_population_by_country.csv')
df
```

```python
df = df.drop(columns = ['Code'])
df
```

```python
df.rename(columns = {'Entity': 'Country', 'Population (historical)': 'Population'}, inplace = True)
```

```python
list_countries = df['Country'].unique()
list_countries
```

```python
# Set list of countries in G20
list_countries_g20 = ["Germany", "France", "Italy", "Argentina", "Australia", "Brazil", "Canada", "China",
"India", "Indonesia", "Japan", "Mexico", "Russia", "Saudi Arabia", "South Africa", "South Korea", "Turkey",
"United Kingdom", "United States"]
```

```python
df_filtered = df[df["Country"].isin(list_countries_g20)]
df_filtered
```

```python
df_filtered = df_filtered[df_filtered['Year'] > 1950]
df_filtered
```

```python
df_filtered.to_csv('population_by_country_from_1950.csv', index = False)
```
